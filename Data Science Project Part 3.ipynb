Section 1
1.a. The authors grew single-strain and mixed bacterial cultures of E. coli strains that had been genetically transformed to carry different antibiotic resistance markers. These genetic differences allowed them to distinguish competing populations during coculture using red or green fluorescent proteins. The growth of each well was monitored automatically using a plate reader, which recorded optical density (OD), OD595, over time to estimate cell population density. This dataset was ideal for testing the authors’ hypothesis because OD measurements directly reflect microbial growth rate and carrying capacity, the two key variables needed for modeling population dynamics. The researchers grew E. coli in both single-strain cultures, where each well contained only one genetically distinct strain, and mixed cultures, where two strains were combined in the same well. This design allowed them to compare growth under both isolation and competition conditions, providing direct data for modeling microbial interactions. The fluorescence data for the individual strains while in mixed culture was reported in two separate tables included with the main dataset.

The features are the time of each observation, the temperature, the strain of each well, the color of each well (green or red), and the well labels for a total of 99 features. The time tells when the observation was taken during the microbial growth, the temperature shows the cells are growing under proper conditions, the strain indicates the type of cell growing in the wells, the color indicates the relative population of both strains during mixed growth, and the 95 well labels indicate where the observation is taken. 

In the main dataset there are 6,365 observations. Each observation corresponds to a single well at a single time point, where the optical density (OD) of the culture was measured. Each well contained either a single strain or a pair of strains (in mixed conditions), and OD was recorded repeatedly over time to capture growth curves. This means the total number of observations equals the number of wells multiplied by the number of times measured. Two other tables are included in the file with the main data set to identify individual strain growth in the mixed culture. These tables have the same observations and features, but they do not include the strain and color features as a table.

1.b. The data set collected with the Tecan microplate reader provides detailed, time-resolved measurements of OD and fluorescence intensity for microbial cultures growing under controlled conditions. Each well on the 96-well plate contains a different strain or experimental condition, and the instrument records OD values at regular time intervals, producing complete growth curves for each strain. 

This information is directly related to the paper’s research goal, which was to determine whether growth in mixed microbial cultures can be predicted from the individual growth dynamics of each strain measured in isolation. The Tecan data captures how quickly each strain grows, how long it takes to reach saturation, and what its maximum density is. These parameters are essential inputs for mathematical models that describe microbial competition and predict relative fitness in mixed environments.

A research question that can be asked is how does the relative abundance of two E. coli strains change over time in mixed culture compared to their individual growth rates in isolation?

Figure 1 was created from single-strain growth data obtained using a Tecan microplate reader, the same type of data found in the dataset. Panel A plots mean cell density (OD) on the y-axis against time (hours) on the x-axis. The open circles represent measured OD values across replicates. The solid black line is a polynomial fit used to smooth experimental data. The dashed line shows an exponential model fit, representing the theoretical exponential growth phase. Panel B plots the growth rate (dN/dt ÷ N) on the y-axis against time (hours) on the x-axis, calculated by differentiating the fitted curve from Panel A. The peak growth rate corresponds to the time of maximum exponential growth, tₘₐₓ. This analysis was done by taking the logarithm of OD values and performing a linear regression during the exponential phase to estimate the growth rate (r). 

The authors included this figure to show the step-by-step approach they used to analyze their experimental growth data. It demonstrates how OD readings were turned into measurable biological values (growth rate, lag time, and carrying capacity) that were later used in the mixed-culture competition models. Overall, this figure lays out the data-analysis method that supports the predictive modeling in later figures. 

We can conclude that the exponential model fits well during the early and mid-log phases of growth but diverges slightly as the culture approaches saturation. Also, the growth-rate curve (Panel B) confirms that microbial growth follows a predictable pattern, with a rapid increase to a peak rate followed by a gradual slowdown. Finally, these results show that the exponential model works well for describing microbial growth in its active phase, giving reliable parameters for later prediction.

Section 3.3 Visualization 1:
For this feature, I chose to make a scatterplot of OD at cycle 10 across all wells. A scatterplot is appropriate here because we have one numerical value (OD at a single time point) for each categorical well label, and we want to see how those values vary from well to well. Plotting OD on the y-axis and well ID on the x-axis lets us quickly see whether most wells have similar OD or whether there are large differences or outliers at this early point in the experiment.

```python
# Visualization 1
import seaborn as sns
import pandas as pd

# OD at cycle 10, wells only (skip time/temp rows)
cycle_10_df = df_od.iloc[2:][[10]].reset_index()
cycle_10_df.columns = ["Well", "OD_cycle_10"]

sns.scatterplot(data=cycle_10_df, x="Well", y="OD_cycle_10")
```

Section 3.3 Visualization 1:
The scatterplot shows that most wells have OD values between roughly 0.63 and 0.75, with a few wells noticeably lower (including one near 0.45 and one close to 0.10). This agrees with the describe output from 3.2, which reported a mean OD ≈ 0.675, std ≈ 0.082, a minimum of 0.10, and a maximum of 0.77 for cycle 10. The tight vertical band of points around the mean and the presence of a small number of low outliers visually confirm the numerical summary: most wells are already at fairly high density at cycle 10, but a few lag behind.

This visualization relates to the paper’s hypothesis about predicting mixed-culture growth from growth-curve data because it illustrates how similar many wells are at a given time point, despite being different strain or culture-type combinations. The fact that OD at cycle 10 is already high and relatively consistent across wells supports the idea that growth follows a predictable pattern that can be captured by simple models, as in the figures where the authors fit growth curves and competition models to OD data over time.

Section 3.3 Visualization 2:
For this second feature I made a scatterplot of OD vs. cycle number for well B8. This is appropriate because we’re now looking at how a single culture changes over time, so cycle number plays the role of a time axis and each point is one measurement of OD in that well. A scatterplot lets us see the overall trend in B8’s growth curve—whether OD is rising, flat, or declining across the 67 cycles—and whether there is much noise or any sudden jumps that might indicate measurement issues.

```python
# Visualization 2
import seaborn as sns
import pandas as pd

# assume OD cycle columns start after the first 3 metadata columns
cycle_columns = df_od.columns[3:]

# OD values for well B8 across cycles
b8_values = pd.to_numeric(df_od.loc["B8", cycle_columns], errors="coerce")

b8_df = pd.DataFrame({
    "Cycle": cycle_columns.astype(int),
    "OD": b8_values.values
})

sns.scatterplot(data=b8_df, x="Cycle", y="OD")
```

Section 3.3 Visualization 2:
The scatterplot shows that OD in red well B8 starts fairly high (around 0.71) and then gradually declines to about 0.65 across the experiment, forming a downward trend rather than the classic sigmoidal growth curve. This does not match the behavior we would expect if we were seeing the full growth phase; instead, it looks like B8 is already near carrying capacity or even in a slight decline when the recorded cycles begin. In 3.2, when I tried to apply describe() to B8, pandas treated the values as categorical and returned count, unique, top, and freq instead of numeric statistics, so that summary was not very informative. The scatterplot exposes the actual numeric pattern and makes it clear that B8’s OD is not increasing over time.
This example is important because it does not strongly support the authors’ hypothesis on its own. The paper’s central idea is that mixed-culture growth can be predicted from single-strain growth curves, but the odd behavior of a single red-strain well like B8 (already high and slowly declining) would be misleading if taken in isolation. Instead, it highlights the importance of having many replicate wells and analyzing patterns across them rather than relying on one example. The authors’ models are fit to average behavior and many replicates, which smooth out wells like B8 that may be in a slightly different phase or affected by noise. This well is a good reminder that individual trajectories can deviate from the “typical” growth curve, and that robust predictions depend on replicates and summary statistics, not single outlier wells.
